```python
def isPrime(n): 
    n=int(n)
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
```

```python
calculation_dimension=10
max_x_value=7
figure(figsize(10,10))

y=lambda x: x**2 #PARABOLA

plot(x,y(x)) #PLOT THE PARABOLA

myrange=arange(-calculation_dimension,calculation_dimension)
myrange=myrange[(myrange<=-2) | (myrange>=+2)]

#Plot the lines
for i1 in myrange:
    for i2 in myrange:
        plot([i1,i2],[y(i1),y(i2)],"-k",alpha=.1)

        
#Plot the labels  
for i in arange(-max_x_value+1,max_x_value+1):
    plot(i,y(i),".")
    if i>0:
        text(i,y(i),"%    d"%i, ha="left")
    if i<0:
        text(i,y(i),"%d    "%i, ha="right")

#Plot a label for each prime number or a black dot.
for k in range(1,int(y(max_x_value)+1)):
    if not isPrime(k):
        plot(0,k,"ok")
    else:
        text(0,k,"%d"%k, ha="center", va="center")

ylim(0,y(max_x_value))
xlim(-max_x_value,max_x_value)
savefig("primes.png")
```
